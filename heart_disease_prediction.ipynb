```python
from scipy import stats
z = np.abs(stats.zscore(data))
print(z)
```

```python
threshold = 3
print(np.where(z > 3))# The first array contains the list of row numbers and second array respective column numbers
```

```python
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
data = data[(z < 3).all(axis=1)]
data.shape
```

```python
data = data[~((data < (Q1 - 1.5 * IQR)) |(data > (Q3 + 1.5 * IQR))).any(axis=1)]
data.shape
```

```python
del data["fbs"]
data.head()
```

```python
from sklearn.preprocessing import StandardScaler
import dask.dataframe as dd
standardScaler = StandardScaler()
dataset = pd.get_dummies(data, columns = ['sex', 'cp','restecg', 'exang', 'slope', 'ca', 'thal'])# creating dummy variable
columns_to_scale = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak'] # we have taken these columns for scale down
dataset[columns_to_scale] = standardScaler.fit_transform(dataset[columns_to_scale])
```

```python
dataset.head()
```

```python
dataset.describe()
```

```python
#sns.pairplot(data , hue="target", height=3, aspect=1);
```

```python
y = dataset['target']
X = dataset.drop(['target'], axis = 1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state= 5)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
knn_classifier = KNeighborsClassifier(n_neighbors = 5)
knn_classifier.fit(X_train, y_train)
score=cross_val_score(knn_classifier,X_train,y_train,cv=10)
y_pred_knn = knn_classifier.predict(X_test)
accuracy_score(y_test, y_pred_knn)
```

```python
score.mean()
```

```python
knn_classifier  = KNeighborsClassifier(algorithm='auto', leaf_size=30, metric='minkowski',
 metric_params=None, n_jobs=1, n_neighbors=5, p=1,
 weights='uniform')
knn_classifier.fit(X_train, y_train)
score=cross_val_score(knn_classifier,X_train,y_train,cv=10)
y_pred_knn = knn_classifier.predict(X_test)
accuracy_score(y_test, y_pred_knn)
```

```python
score=cross_val_score(knn_classifier,X_train,y_train,cv=10)
score.mean()
```

```python
#confustion metrix
cm = confusion_matrix(y_test, y_pred_knn)
plt.title('Heatmap of Confusion Matrix', fontsize = 15)
sns.heatmap(cm, annot = True)
plt.show()
```

```python
print(classification_report(y_test, y_pred_knn))
```
